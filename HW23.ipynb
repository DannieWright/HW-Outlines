## 8.2 Differential Equations With More Than One Variable

This is having _simultaneous differential equations_ wher the derivatives of each variable can depend on any or all of the other variables, as well as the original independant variable. The general form for two first-order simultaneous differential equations is

$\frac{dx}{dt} = f_x (x, y, t), \\
 \frac{dy}{dt} = f_y (x, y, t).$
 
Or it can be written in vector form for an arbitrary number of variables as

$\frac{d\vec{r}}dt = \vec{f}(\vec{r}, t).$

where $\vec{r} = (x, y, \dots)$ and $\vec{f} = (f_x(\vec{r}, t), f_y(\vec{r}, t), \dots)$.

Then we simply do what we did before, but our k-values are arrays.

```python
#example 8.5
from math import sin
from numpy import arange, array
from pylab import plot, xlabel, ylabel, show

def f (r, t):
    x = r[0]
    y = r[1]
    fx = x * y - x
    fy = y - x * y + sin (t)**2
    return array ([fx, fy], float)

a = 0 #start of interval
b = 10 #end of interval
N = 1000 #number of steps
h = (b - a) / N #step size

tPoints = arange (a, b, h)
xPoints = []
yPoints = []

r = array ([1.0, 1.0], float)

for t in tPoints:
    xPoints.append (r[0])
    yPoints.append (r[1])
    k1 = h * f (r, t)
    k2 = h * f (r + 0.5 * k1, t + 0.5 * h)
    k3 = h * f (r + 0.5 * k2, t + 0.5 * h)
    k4 = h * f (r + k3, t + h)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6
    
plot (tPoints, xPoints)
plot (tPoints, yPoints)
xlabel ("t")
show ()
```
